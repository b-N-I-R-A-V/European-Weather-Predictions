# 2 Scaling

### 2.1 Subset data

```python
# All columns except for data and month will be scalled
df2 = df.drop(columns = ['DATE','MONTH'])
```

### 2.2 Check for (and handle) missing values

```python
df2.isnull().sum().loc[df2.isnull().sum()==0]
```

There are no missing values for all the columns.

### 2.3 Scaling

```python
# Scaling using StandardScaler from sklearn.preprocessing
scaler = StandardScaler()
```

```python
#Creating new df with scaled data

df_scaled = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)
df_scaled.head()
```

```python
# Adding Month and Date back to the scaled data
df_scaled['month'] = df['MONTH']
df_scaled['date'] = df['DATE']
```
